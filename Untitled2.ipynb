```python
import pandas as pd
import numpy as np

df=pd.read_csv('gross.csv')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(1,3))
cols_to_scale = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper','Delicassen']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
print(df.head())
print(df.describe())
```

```python
df['class'].value_counts()
```

```python
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from collections import Counter
X = df.drop("class", axis=1)
y = df["class"]
print(Counter(y))
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
print(Counter(y_res))
```

```python
df_res = pd.concat([X_res, y_res], axis=1)
n= df_res.duplicated()
print(n.value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
class KNNClassifier:
    def __init__(self, k=5):
        self.k = k
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
    def predict_one(self, x):

        X_train_np = self.X_train.to_numpy() if hasattr(self.X_train, 'to_numpy') else self.X_train
        x_np = x.to_numpy() if hasattr(x, 'to_numpy') else x

        distances = np.sqrt(np.sum((X_train_np - x_np)**2, axis=1))
        k_indices = np.argsort(distances)[:self.k]
        k_neighbor_labels = self.y_train.iloc[k_indices] if hasattr(self.y_train, 'iloc') else self.y_train[k_indices]
        most_common = Counter(k_neighbor_labels).most_common(1)
        return most_common[0][0]
    def predict(self, X):
        return np.array([self.predict_one(X.iloc[i]) for i in range(len(X))])
    def score(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == y)
```

```python
knn_scratch = KNNClassifier(k=5)
knn_scratch.fit(X_train, y_train)
train_acc = knn_scratch.score(X_train, y_train)
test_acc = knn_scratch.score(X_test, y_test)
print(f"KNN (Scratch) - Train Accuracy: {train_acc:.3f}, Test Accuracy: {test_acc:.3f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score
y_pred = neigh.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
